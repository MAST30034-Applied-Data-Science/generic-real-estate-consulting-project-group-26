```python
from urllib.request import urlretrieve
import requests
from bs4 import BeautifulSoup
from urllib.request import urlopen
import pandas as pd
```

```python
URL = "https://www.matthewproctor.com/full_australian_postcodes_vic"
bs_object = BeautifulSoup(requests.get(URL).text, "html.parser")
```

```python
table = bs_object.find("table", {"class": "table table-striped"})

# The attribute we want is located at these index in the table in the website
POSTCODE_INDEX = 1
NAME_INDEX = 2
SA2_CODE_INDEX = 18
SA2_NAME_INDEX = 19
LGA_INDEX = 29
suburb_list = []
for row in table.findAll('tr')[1:]:    
    columns = row.find_all('td')
    postcode = columns[POSTCODE_INDEX].text.strip()
    name = columns[NAME_INDEX].text.strip()
    LGA = columns[LGA_INDEX].text.strip()
    SA2_code = columns[SA2_CODE_INDEX].text.strip()
    SA2_name = columns[SA2_NAME_INDEX].text.strip()
    suburb_list.append([postcode, name, LGA, SA2_code, SA2_name])
```

```python
suburb_df = pd.DataFrame(suburb_list, columns = ["Postcode","Suburb","LGA", "SA2_code", "SA2_name"])
```
