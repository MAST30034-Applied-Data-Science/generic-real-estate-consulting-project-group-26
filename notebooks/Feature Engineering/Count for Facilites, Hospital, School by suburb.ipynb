```python
import numpy as np
import pandas as pd
```

```python
facilites  = pd.read_csv('../../data/curated/Facilites_list.csv')
facilites
```

```python
facilites_suburb = facilites.groupby(['Suburb/Town'], as_index=False)['Facility ID'].count()
facilites_suburb.columns = ['Suburb','Facilites']
facilites_suburb.to_csv('../../data/curated/facilites_suburb.csv')
print(facilites_suburb)
```

```python
hospital = pd.read_csv('../../data/raw/hospitals_vic.csv')
hospital
```

```python
hospital_suburb = hospital.groupby(['Suburb'], as_index=False)['Hospital name'].count()
hospital_suburb.columns = ['Suburb','Hospitals']
hospital_suburb.to_csv('../../data/curated/hospital_suburb.csv')
print(hospital_suburb)
```

```python
school = pd.read_csv('../../data/raw/SchoolLocation.csv')
school
```

```python
school_suburb = school.groupby(['Address_Town'], as_index=False)['School_Name'].count()
school_suburb.columns = ['Suburb','Schools']
school_suburb.to_csv('../../data/curated/school_suburb.csv')
print(school_suburb)
```
