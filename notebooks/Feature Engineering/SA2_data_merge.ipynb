```python
import pandas as pd
```

### Read data and merge

```python
rent=pd.read_csv("../../data/curated/select_rent.csv")
```

```python
suburb=pd.read_csv("../../data/curated/suburb.csv")
```

```python
growth=pd.read_csv("../../data/curated/suburb_growth_5_years.csv")
```

```python
growth.head()
```

```python
growth.shape
```

```python
properties = pd.read_csv('../../data/curated/properties.csv')
```

```python
properties["Suburb"]=properties["suburb"].str.upper()  # create common attribute for merging
```

```python
properties.shape
```

```python
properties.head()
```

```python
properties = properties[["name","type","prop_lat","prop_long","Beds","Baths","Parking","weekly_cost","Suburb"]]
```

```python
property_count = properties.groupby(['Suburb']).count()[['name']]  #calculate the number of properties in each suburb
property_count.columns = ['num_property']
```

```python
property_count.head()
```

```python
#property_count.to_csv('../data/curated/property.csv')
```

```python
property_count.shape
```

```python
merge=pd.merge(growth,rent,on="Suburb",how="inner")  #merge data on suburb, apply inner jpin to avoid null data, 
```

```python
merge.head()
```

```python
merge.shape
```

```python
merge=pd.merge(merge,property_count,on="Suburb",how="inner")
```

```python
merge=pd.merge(merge,suburb,on="Suburb",how="inner")
```

```python
merge.shape
```

```python
merge.head()
```

```python
merge["% rental price Growth in 5 years"].head()
```

### Select useful data attribute

```python
select=merge[["Suburb","SA2_name","SA2_code","Postcode","LGA","Mar 2021 Median","% rental price Growth in 5 years",\
              "% income Growth in 5 years","% population Growth in 5 years","num_property"]]  
```

```python
select.head()
```

### calculate the mean grwoth rate for each SA2

```python
select1=select[["Suburb","SA2_name","Mar 2021 Median","% rental price Growth in 5 years","% income Growth in 5 years",\
                "% population Growth in 5 years"]].groupby("SA2_name").mean()
```

```python
select1.head()
```

### calculate the number of property in each SA2 

```python
select2=select[["SA2_name","num_property"]].groupby("SA2_name").sum()
```

```python
select2.head()
```

### Merge data on SA2

```python
merge_selected=pd.merge(select1,select2,on="SA2_name",how="inner")
```

```python
merge_selected.head()
```

```python
merge_selected.to_csv("../../data/curated/map_data.csv")
```
