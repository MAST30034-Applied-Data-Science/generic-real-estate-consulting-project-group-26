```python
# import required libries
import pandas as pd
```

### Read data and merge

```python
# read in dataset
rent=pd.read_csv("../../data/curated/select_rent.csv")
```

```python
rent.head()
```

```python
# read in dataset
suburb = pd.read_csv("../../data/raw/suburb.csv")
```

```python
suburb.head()
```

```python
# read in dataset
growth = pd.read_csv("../../data/curated/suburb_growth_5_years.csv")
```

```python
growth.head()
```

```python
properties = pd.read_csv('../../data/curated/properties.csv')
```

```python
# create common attribute for merging
properties["Suburb"] = properties["suburb"].str.upper()  
```

```python
properties.head()
```

```python
#Select required attributes
properties = properties[["name", "type", "prop_lat", "prop_long",
                         "Beds", "Baths", "Parking", "weekly_cost", "Suburb"]]
```

```python
#calculate the number of properties in each suburb
property_count = properties.groupby(['Suburb']).count()[['name']]  
property_count.columns = ['num_property']
```

```python
merge=pd.merge(growth, rent, on="Suburb", how="inner")  #merge data on suburb, apply inner jpin to avoid null data, 
```

```python
merge.head()
```

```python
merge = pd.merge(merge, property_count, on="Suburb", how="inner")
```

```python
merge = pd.merge(merge, suburb, on="Suburb", how="inner")
```

```python
merge.head()
```

```python
merge["% rental price Growth in 5 years"].head()
```

### Select useful data attribute

```python
# Select useful data attribute
select=merge[["Suburb", "SA2_name", "SA2_code", "Postcode", "LGA",
              "Mar 2021 Median", "% rental price Growth in 5 years",
              "% income Growth in 5 years", "% population Growth in 5 years", "num_property"]]  
```

```python
select.head()
```

### calculate the mean grwoth rate for each SA2

```python
# calculate the mean grwoth rate for each SA2
select1=select[["Suburb","SA2_name","Mar 2021 Median",
                "% rental price Growth in 5 years",
                "% income Growth in 5 years",
                "% population Growth in 5 years"]].groupby("SA2_name").mean()
```

```python
select1.head()
```

### calculate the number of property in each SA2 

```python
# calculate the number of property in each SA2 
select2=select[["SA2_name","num_property"]].groupby("SA2_name").sum()
```

```python
select2.head()
```

### Merge data on SA2

```python
# merge dataframes
merge_selected = pd.merge(select1, select2, on="SA2_name", how="inner")
```

```python
merge_selected.head()
```

```python
merge_selected.to_csv("../../data/curated/map_data.csv")
```
